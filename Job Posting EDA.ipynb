# What salary ranges are common? 

```python
data['Salary'].unique()
```

```python
data['Salary'].value_counts()
```

```python
data = pd.DataFrame({
    'Salary': ['No salary', '70000.00-120000.00', '110000.00-140000.00', '250000.00-275000.00', '80000.00-100000.00', '140000.00-180000.00', '80000.00-150000.00', '120000.00-160000.00', '100000.00-165000.00', '85000.00-175000.00', '80000.00-90000.00', '70000.00-90000.00']
})

# Extract numeric values from salary ranges and take the average
data['Salary'] = data['Salary'].str.extractall(r'(\d+\.\d+)').astype(float).groupby(level=0).mean()

# Sort the DataFrame by the 'Salary' column
data = data.sort_values(by='Salary')

# Plot the horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 6))

ax.barh(data['Salary'].index, data['Salary'], color='skyblue')

ax.xaxis.set_tick_params(pad=10)
ax.yaxis.set_tick_params(pad=20)

plt.title('Salary Ranges')
plt.xlabel('Average Salary')
plt.ylabel('Index')
plt.show()
```

# What are the ranges for job application numbers in relation to date posted?

```python
data['Published date'] = pd.to_datetime(data['Published date'], errors='coerce')

applications_by_date = data.groupby('Published date')['Application count'].sum()

plt.figure(figsize=(12, 6))
applications_by_date.plot(marker='o', linestyle='-')
plt.title('Job Applications Over Time Based on Published Date')
plt.xlabel('Published Date')
plt.ylabel('Total Applications')
plt.grid(True)
plt.show()
```

# What are the sector type categories? 

```python
data['Sector'].value_counts()
```

```python
sector_counts = data['Sector'].value_counts()

plt.figure(figsize=(12, 6))
sector_counts.plot(kind='bar', color='skyblue')
plt.title('Counts of Job Postings by Sector')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
words_to_exclude = ['IT', 'Technology', 'Software Development', 'IT services', 'IT consulting']

filtered_data = data[~data['Sector'].str.contains('|'.join(words_to_exclude), case=False)]

filtered_sector_counts = filtered_data['Sector'].value_counts()

plt.figure(figsize=(12, 6))
filtered_sector_counts.plot(kind='bar', color='skyblue')
plt.title('Counts of Job Postings by Sector (Excluding IT-related Sectors)')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 # Use regular expression to find specific tools, experience, and location type in job descriptions

```python
data = pd.DataFrame({
    'Description': [
        "Python",
        "SQL",
        "Tableau",
        "Power bi",
        "Git",
        "Mongo DB",
        "PySpark",
        "AWS",
        "Azure",
        '2 years of experience',
        '5+ years of experience',
        '3-4 years of experience',
        '7-10 years of experience',
        'WFH',
        'remote',
        'on site',
        'location',
        'hybrid'
    ]
})

pattern_a = r'(\d+)\s*[-+]*\s*(?:year|yr|experience|exp)'
pattern_b = r'\b(Python|SQL|Tableau|Power bi|Git|Mongo DB|PySpark|AWS|Azure)\b'
pattern_c = r'\b(WFH|home|remote|site|location|hybrid)\b'

data['ExperienceYears'] = data['Description'].str.extract(pattern_a, flags=re.IGNORECASE)
data['Tools'] = data['Description'].str.extract(pattern_b, flags=re.IGNORECASE)
data['Location'] = data['Description'].str.extract(pattern_c, flags=re.IGNORECASE)

data['ExperienceYears'] = pd.to_numeric(data['ExperienceYears'], errors='coerce')

print(data)

```
