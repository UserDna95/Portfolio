# Amount of revenue growth per industry 

```python
revenue = df['Revenue growth'].head(12)
industry = df['Industry'].head(12)

fig, ax= plt.subplots(figsize =(16, 9))

ax.barh(industry, revenue)

ax.xaxis.set_tick_params(pad = 10)
ax.yaxis.set_tick_params(pad = 20)

plt.show()
```

# Top 10 companies with highest revenue

```python
df_sorted = df.sort_values(by='Revenue (USD millions)', ascending=False)

df_top10 = df_sorted.head(10)

ax = df_top10.plot(kind='scatter', y='Revenue (USD millions)', x='Name', figsize=(12, 8))

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

ax.set_yticks(range(len(df_top10)))
ax.set_yticklabels(df_top10.index)

plt.subplots_adjust(left=0.2, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

# Revenue growth rate for top 10 companies

```python
data = {
    'Name': ['Walmart', 'Amazon', 'Exxon Mobil', 'Apple', 'UnitedHealth Group', 'CVS Health', 'Berkshire Hathaway', 'Alphabet', 'McKesson Corporation', 'Chevron Corporation'],
    'Revenue (USD millions)': [1000000, 1200000, 800000, 1500000, 900000, 1100000, 1300000, 950000, 850000, 1400000],
    'Revenue growth': [0.15, 0.10, 0.20, 0.08, 0.12, 0.18, 0.09, 0.14, 0.13, 0.11]
}

df = pd.DataFrame(data)

df['Revenue-to-Growth Ratio'] = df['Revenue (USD millions)'] / df['Revenue growth']

df_sorted = df.sort_values(by='Revenue-to-Growth Ratio', ascending=False)

top_10_companies = df_sorted.head(10)

top_10_companies['Revenue-to-Growth Ratio'] = top_10_companies['Revenue (USD millions)'] / top_10_companies['Revenue growth']

print(top_10_companies[['Name', 'Revenue-to-Growth Ratio']])

plt.figure(figsize=(12, 8))
plt.bar(top_10_companies['Name'], top_10_companies['Revenue-to-Growth Ratio'], color='darkblue')
plt.title('Top 10 Companies based on Revenue-to-Growth Ratio')
plt.xlabel('Name')
plt.ylabel('Revenue-to-Growth Ratio')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
